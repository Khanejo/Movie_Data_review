```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualize the distribution of numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64'])
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Visualize the correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Visualize the distribution of categorical features
categorical_features = df.select_dtypes(include=['object'])
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=feature, order=df[feature].value_counts().index)
    plt.xticks(rotation=90)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
cat_col = categorical_features.columns
num_col = numerical_features.columns

# Impute numerical NaNs with mean for each numerical column
for col in num_col:
    df[col].fillna(df[col].mean(), inplace=True)

# Impute categorical NaNs with mode for each categorical column
for col in cat_col:
    df[col].fillna(df[col].mode().iloc[0], inplace=True)

df
```

```python
# Split dataset to train_test
from sklearn.model_selection import train_test_split

Y = df['imdb_score']
X = df.drop(['imdb_score'],1)

```

```python
from sklearn.preprocessing import LabelEncoder
import category_encoders as ce
# Encode low cardinality columns using Label Encoding
for col in cat_col:
    if len(X[col].unique()) <= 5:
        label_encoder = LabelEncoder()
        X[col] = label_encoder.fit_transform(X[col])

# Identify high cardinality columns
high_cardinality_columns = [col for col in cat_col if len(X[col].unique()) > 5]

# Encode high cardinality columns using Binary Encoding
encoder = ce.BinaryEncoder(cols=high_cardinality_columns)
X_encoded = encoder.fit_transform(X[high_cardinality_columns])

# Concatenate
X = pd.concat([X, X_encoded], axis=1)
```

```python
X = X.drop(high_cardinality_columns, axis=1)
X
```

```python
from sklearn.preprocessing import RobustScaler
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=32)

# Initialize the Robust Scaler
scaler = RobustScaler()

# Fit the scaler on the training data and transform it
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
X.describe()
```

```python
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import mean_squared_error

reg = RandomForestRegressor(n_estimators=100, random_state=0)
reg.fit(X_train_s, y_train)
mse = mean_squared_error(y_test, reg.predict(X_test_s))
mse
```

```python
from sklearn.ensemble import GradientBoostingRegressor 

reg = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=0)
reg.fit(X_train_s, y_train)
mse = mean_squared_error(y_test, reg.predict(X_test_s))
mse
```

```python
from sklearn.linear_model import Lasso

reg = Lasso(alpha=1.0)
reg.fit(X_train_s, y_train)

y_pred = reg.predict(X_test_s)
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
from sklearn.svm import SVR


reg = SVR(kernel='linear', C=1.0)
reg.fit(X_train_s, y_train)


y_pred = reg.predict(X_test_s)
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
from tensorflow import keras
from tensorflow.keras import layers


model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train_s.shape[1],)),
    layers.Dropout(0.5),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train_s, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)

y_pred = model.predict(X_test_s)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
```

#### I'd prefer to use GradientBoostingRegressor to be used in the production environment owing to the metrics Root mean square score. Also, GradientBoostingRegressor is flexible, extendable.
